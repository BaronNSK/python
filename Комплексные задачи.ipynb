```python
# числовая угадайка, где угадывает человек
from random import *

print('Добро пожаловать в числовую угадайку')
print('Копмьютер загадывает случайное число в границах, указанных вами')

# подходит ли ответ
def is_valid(x, edge_l, edge_r):
    if x.isdigit() and edge_l <= int(x) <= edge_r:
        return True
    else:
        return False

# расчёт минимального количества ходов, гарантирующих угадывание
def min_garant(x):
    c = 0
    if x == 0:
        return 1
    elif x == 1 or x == 2:
        return 2
    if x % 2 == 1 and len(list(range(x))) == 3:
        c += 1
    while x != 1:
        if x % 2 != 0:
            x += 1
        x //= 2
        c += 1
    return c

# границы
def edges(l, r):
    if not l.isdigit() or not r.isdigit():
        while not l.isdigit() or not r.isdigit():
            print('Границы должны быть числами')
            l, r = input(), input()
            if l.isdigit() and r.isdigit():
                break
    if int(l) > int(r):
        l, r = r, l
    return int(l), int(r)

# ввод границ
def play():
    global edge_l, edge_r, x, c
    print('Введите числа, которые будут границами, после каждого числа нажмите ENTER')
    edge_l, edge_r = input(), input()
    edge_l, edge_r = edges(edge_l, edge_r)
    x = randint(edge_l, edge_r)
    c = 0

def yes_or_no(x):
    global again
    while again != 'да' and again != 'нет':
        again = input().lower()
        if again == 'да':
            play()
        elif again == 'нет':
            break
        else:
            print('Введите "Да" или "Нет"')

print('Введите числа, которые будут границами, после каждого числа нажмите ENTER')
edge_l, edge_r = input(), input()
edge_l, edge_r = edges(edge_l, edge_r)
x = randint(edge_l, edge_r)
c = 0
again = ''
print('Если долго не можете угадать число, введите "Сдаюсь"')

while True:
    if again == 'нет':
        break
    again = ''
    c += 1
    print(f'Введите число от {edge_l} до {edge_r}')
    answer = input().lower()
    if answer == 'сдаюсь':
        print(f'Компьютер загадал число {x}')
        print('Хотите сыграть ещё? (Да/Нет)')
        yes_or_no(again)
    if again == 'нет':
        break
    if not is_valid(answer, edge_l, edge_r):
        print(f'А может быть все-таки введем целое число от {edge_l} до {edge_r}?', sep = '')
        continue
    else:
        answer = int(answer)
    if answer < x:
        print('Ваше число меньше загаданного, попробуйте ещё раз')
    elif answer > x:
        print('Ваше число больше загаданного, попробуйте ещё раз')
    else:
        print('Вы угадали, поздравляем!')
        print(f'Ваше число попыток = {c}')
        print(f'Минимальное число попыток, ГАРАНТИРУЮЩИХ угадывание загаданного числа в указанных вами границах = {min_garant(edge_r - edge_l)}')
        print('(это число высчитывается по специальной формуле)')
        print('Хотите сыграть ещё? (Да/Нет)')
        yes_or_no(again)
                
print('Спасибо за игру.')
```
